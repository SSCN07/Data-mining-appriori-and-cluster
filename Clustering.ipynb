```python
import numpy as np
def k_means_cluster(x,k=[3,3.9,8],iters=20):
    
    for _ in range(iters):
        p1, p2, p3= [],[],[]
        d=[0,0,0]
        for i in range(len(x)):
            for j in range(len(k)):
                d[j]=abs(x[i]-k[j])#for distance in absolute value

            mind=d.index(min(d)) #index of the minimum value in a list d.# Find which centroid is closest
            if mind==0:
                p1.append(x[i])
            if mind==1:
                p2.append(x[i])
            if mind==2:
                p3.append(x[i])
        k[0]= np.mean(p1)
        k[1]= np.mean(p2)
        k[2]= np.mean(p3)
    return k,p1,p2,p3
```

```python
k,p1,p2,p3= k_means_cluster(x)
```

```python
def clusterM(X,K=3,iters=10):
    init_k = np.random.rand(K, X.shape[1])
    m=X.shape[0]
    v=[]
    for i in range(iters):
        v = np.zeros(m, dtype=int)
        for i, point in enumerate(X):
            distances = np.linalg.norm(point - init_k, axis=1)#for distance in absolute value
            v[i] = np.argmin(distances) # np.argmin() for array index smallest value 
            v=np.array(v)
        for i in range(K):#
            cluster_points = X[v == i]
            init_k[i] = np.mean(cluster_points, axis=0)

    return init_k, v    
```

```python
clusterM(X,K=3,iters=10)
```
